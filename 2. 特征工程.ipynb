# 1. 特征工程

## 1.1 什么是特征工程     
定义：将数据转换为能更好地表示潜在问题的特征，从而提高机器学习性能。
     
解释：    
转换过程    
1. 为机器学习pipline进行服务的（更好地表示潜在问题）。     
2. 数据清洗（丢失、冗余）、数据组织（行、列）、数据转换。
      
特征   
列（属性）：普通 + 隐藏（建模）  
      
标准评估指标     
baseline、阈值 
一定要考虑应用场景（股票价格预测：监督学习/回归预测问题）  


## 1.2 特征理解
1. 识别结构化数据与非结构化数据
2. 定量数据与定性数据
3. 数据的4个等级：定类等级、定序等级、定距等级、定比等级
4. 探索性数据分析和数据可视化
5. 描述性统计

## 1.5 特征构建
为数据集插入新的特征
1. 基于业务背景构建更能表达问题的特征（价格单日变化幅度、单周成交量）---- 基于统计信息
2. 模型可以有效利用的特征：常见的有one-hot编码、{0，1}
2. 实体匹配（淘宝销量预测利用客户京东购物数据）    
3. 空间投影（价格单日变化幅度-->某个空间（市场情绪、政策影响）） ---- 领域知识（GNN、NLP）  

场景：事件/趋势相似性发现    
图相似度搜索：常用图编辑距离或者最大共同子图来衡量图的相似度，然而这两个指标的计算复杂度非常高（NP-complete）。

```python
del babaDf['newFeat']
babaDf
```

```python
# 特征工程
def change(column):
    res = [0]
    buyPrice = column[0]
    print(len(column))
    for i in range(1, len(column)):
        res.append((column[i] - buyPrice) / buyPrice)
    return res

# 获取收盘价Close这一列的数据
closeCol = babaDf['Close']
# 构建累计涨跌幅这一新特征
babaChange = change(closeCol)
```

```python
babaDf['cumulativeChange'] = babaChange
babaDf
```
